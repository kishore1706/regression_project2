```python
#create a dataframe with 3 customers and 3 products and their prices and customer name be cus1,cus2,cus5
data={"customer":["cus1","cus2","cus5"],"product":["prod1","prod2","prod3"],"price":[100,200,300]}
df=pd.DataFrame(data)
#print(df.groupby("customer")["price"].sum())
print(df.pivot_table(index="customer",values="price",aggfunc="min"))
```

```python
#create two dataframes to perform merge and join operations
data1=pd.DataFrame({"customer":["cus1","cus2","cus5"],"product":["prod1","prod2","prod3"],"price":[100,200,300]})
data2=pd.DataFrame({"customer":["cus1","cus3","cus4"],"product":["prod1","prod2","prod3"],"price":[300,500,600]})

print(pd.merge(data1,data2,on='product'))
print(data1.set_index("customer").join(data2.set_index("customer"),lsuffix="_left",rsuffix="_right"))
```
