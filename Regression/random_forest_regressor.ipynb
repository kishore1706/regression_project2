```python
import pandas as pd
df = pd.read_csv("C:\\Users\\chandini ch\\Downloads\\student_performance_200_rows.csv")
df.head(3)
```

```python
X=df[["Study_Hours","Attendance","Assignments"]]
y=df["Marks"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 30,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators=37,max_depth=10,random_state=42)
rfr.fit(X_train,y_train)
rfr_predict = rfr.predict(X_test)
print(rfr_predict)
```

```python
from sklearn.metrics import mean_absolute_error,root_mean_squared_error,r2_score,mean_absolute_percentage_error
msa = mean_absolute_error(y_test,rfr_predict)
rmse = root_mean_squared_error(y_test,rfr_predict)
r2score = r2_score(y_test,rfr_predict)
error_percentage = mean_absolute_percentage_error(y_test,rfr_predict)
print(msa)
print(rmse)
print(r2score)
print(error_percentage)
```

```python
from sklearn.model_selection import GridSearchCV
rfr = RandomForestRegressor(random_state=42)
params = {'n_estimators':range(1,200),'max_depth':range(2,11)}
grid = GridSearchCV(rfr,params,cv=5,scoring="r2",n_jobs=-1,verbose=2)
grid.fit(X_train,y_train)
print(grid.best_params_)
print(grid.best_score_)
```
