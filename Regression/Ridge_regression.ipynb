```python
import pandas as pd
df = pd.read_csv("C:\\Users\\chandini ch\\Downloads\\student_scores.csv")
df.head(3)
```

```python
X=df[["Study_Hours","Attendance","Assignments"]]
y=df["Marks"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.3,random_state=30)
```

```python
from sklearn.linear_model import Ridge
rid = Ridge(alpha=0.1)
rid.fit(X_train,y_train)
pred_rid = rid.predict(X_test)
```

```python
a=pd.DataFrame({"actual":y_test,"predicted":pred_rid})
print(a)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score
mae = mean_absolute_error(y_test,pred_rid)
print(mae)
print(r2_score(y_test,pred_rid))
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=2.19)
lasso_fit = lasso.fit(X_train,y_train)
lasso_predict = lasso.predict(X_test)
#print(lasso_fit.alpha_)
```

```python
df = pd.DataFrame({"actual":y_test,"predicted":lasso_predict})
print(df)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,lasso_predict))
```
