```python
import pandas as pd
df = pd.read_csv("C:\\Users\\chandini ch\\Downloads\\decision_tree_regressor_100_rows.csv")
print(df.head(4))
df.shape
```

```python
X=df[["Study_Hours","Attendance","Assignments"]]
y=df["Marks"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 30,random_state=50)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_train_scaled = ss.fit_transform(X_train)
X_test_scaled = ss.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error,root_mean_squared_error
for i in range(1,22):
    dtr = DecisionTreeRegressor(max_depth=i,random_state=2,min_samples_split=20)
    dtr.fit(X_train,y_train)
    dtr_predict = dtr.predict(X_test)
    msa = mean_absolute_error(y_test,dtr_predict)
    r2 = root_mean_squared_error(y_test,dtr_predict)
    #print(dtr_predict) 
    print("max_depth",i,"mean_absolute_error",msa,"root_mean_squared_error",r2)
```
