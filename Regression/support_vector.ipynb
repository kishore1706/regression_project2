```python
import pandas as pd
df = pd.read_csv("C:\\Users\\chandini ch\\Downloads\\student_performance_200_rows.csv")
print(df.head(3))
df.shape
```

```python
X = df[["Study_Hours","Attendance","Assignments"]]
y= df["Marks"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=30,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_train_scaled = ss.fit_transform(X_train)
X_test_scaled = ss.transform(X_test)
```

```python
from sklearn.svm import SVR
sv_regressor = SVR()
sv_regressor.fit(X_train_scaled,y_train)
sv_predict = sv_regressor.predict(X_test_scaled)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score
mae = mean_absolute_error(y_test,sv_predict)
r2 = r2_score(y_test,sv_predict)
print("mae",mae)
print("r2_score",r2)
```

```python
new_data = pd.DataFrame({"Study_Hours":[6],"Attendance":[85],"Assignments":[7]})
new_data_scaled = ss.transform(new_data)
new_data_predict = sv_regressor.predict(new_data_scaled)
print(new_data_predict)
```
